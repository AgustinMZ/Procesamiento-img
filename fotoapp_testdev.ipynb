```python
from fotoapp import *

#Link Github: https://github.com/AgustinMZ/Procesamiento-img.git

def mostrar_menu():

    print("\n--- Procesamiento de Imágenes ---\n")

    print("1. Redimensionar imagen para una red social")
    print("2. Ajustar contraste histograma")
    print("3. Aplicar filtro")
    print("4. General boceto")
    print("5. Salir\n")

def main():

    imagen_redimensionada = False

    ruta_guardado = 0
    
    while True:     

        mostrar_menu()

        opción = input("Ingrese una opción: ")

        if opción == "1":
            ruta_imagen = input("Ingrese la ruta de su imagen")
            redsocial = input("Ingrese la red social que desea")
            ruta_guardado = redimensionar_imagen(ruta_imagen, redsocial)
            imagen_redimensionada = True
    
        elif opción == "2":
            if  imagen_redimensionada == False:
                print("Redimensionar una imagen (opción 1) antes de usar esta opción.")
            else:
                ajustar_contraste_histograma(ruta_guardado)
        
        elif opción == "3":
            if  imagen_redimensionada == False:
                print("Redimensionar una imagen (opción 1) antes de usar esta opción.")
            else:
                print("Estos son los filtros disponibles: ")
                print("BLUR, CONTOUR, DETAIL, EDGE_ENHANCE, EDGE_ENHANCE_MORE, EMBOSS, FIND_EDGES, SHARPEN, SMOOTH")
                nombre_filtro = input("Ingrese el filtro que desea")
                aplicar_filtro(ruta_guardado, nombre_filtro)

        elif opción == "4":
            if  imagen_redimensionada == False:
                print("Redimensionar una imagen (opción 1) antes de usar esta opción.")
            else:
                generar_boceto(ruta_imagen, persona=True)

        elif opción == "5":  
            print("Saliendo...") 
            break
        
        #Si el usuario escribe algo diferente a los números de las opciones.
        else: print("Error, escriba una opción.")

#Se inicia el sistema.
main()
```
